## **2次元プロット**


> ### [**グラフ**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)
> ```python
> plt.plot(x, y, options)
> ```
> 引数
> - x: shape=(N,)
> - y: shape=(N,)
>
> options（完全なリストは[ドキュメント](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)参照
> - 第三番目の引数は点や線のフォーマット : "-"は線で結ぶ、"o"はマーカー
> - color : 色 str
> - alpha : 透明度 float (0:完全に透明、1:完全に不透明)

````python
x = np.linspace(-10, 10)

plt.plot(x, np.sin(x), "o-", color="red", alpha=0.4)
plt.show()
````

> ### [**ヒストグラム**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist#matplotlib.pyplot.hist)
> ```python
> plt.hist(x, options)
> ```
> 入力
> - x.shape : (N,)
>
> options
>
> - color : 色 str
> - alpha : 透明度 float (0:完全に透明、1:完全に不透明)

````python
rg = np.random.default_rng()
x = rg.normal(0, 1, 5000)

plt.hist(x, bins=20)
plt.show()
````

> ### [**散布図**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)
> ```python
> plt.scatter(x, y, options)
> ```
> 引数
> - x = (N,)
> - y = (N,)
>
> options
> - marker: 丸"o", バツ"x", ドット"."
> - color: 色 str


````python
rg = np.random.default_rng()
xy = rg.standard_normal(size=(100, 2))

plt.scatter(xy[:,0], xy[:,1], marker="x", alpha=0.8)
````

> ### [**2次元イメージ表示**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html?highlight=imshow#matplotlib.pyplot.imshow)
> ```python
> plt.imshow(x, options)
> ```
> 入力のshape
> - x.shape : (Lx, Ly)（単色）か
> - x.shape : (Lx, Ly, C)（カラー）
>
>Cはカラー画像の場合のチャネル数
>
> options
> - cmap: 単色の場合の色彩
> - interpolation: ぼかし。ぼかしなしは"none"


````python
x_image = rg.uniform(size=(32,32))

plt.imshow(x_image, cmap="gray")
plt.colorbar() # colobar is available
plt.show()
````

````python
x_image = rg.uniform(size=(32,32,3))

plt.imshow(x_image)
plt.show()
````

> ### [**ベクトル場**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html?highlight=quiver#matplotlib.pyplot.quiver)
> ```python
> plt.quiver(x, y, u, v, options)
> ```
> 入力のshape
> - x.shape : (M,) x(縦)方向の格子点
> - y.shape : (N,) y(横)方向の格子点
> - u.shape, v.shape : (N, M) 格子点(n, m)上のベクトル成分 (u, v)

````python
x = np.linspace(-1, 1 ,10)
y = np.linspace(-1, 1, 10)
u, v = np.meshgrid(x, y)

plt.quiver(x, y, u, v, color="blue")
````

> ### [**等高線（中塗りなし）**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html?highlight=contour#matplotlib.pyplot.contour)
> ```python
> plt.contour(X, Y, Z, options)
> ```
> X, Yで指定した格子点上に、Zで指定された等高線を書く
> - X.shape : (M, N)
> - Y.shape : (M, N)
> - Z.shape : (M, N)
>
> options


````python
x = np.linspace(-1, 1 ,10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contour(X, Y, Z)
plt.colorbar()
````

> ### [**等高線（中塗りあり）**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html?highlight=contour#matplotlib.pyplot.contour)
> ```python
> plt.contourf(X, Y, Z, options)
> ```
> contourにfをつけるだけ

````python
x = np.linspace(-1, 1 ,10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contourf(X, Y, Z)
plt.colorbar()
````

### **重ねて描く**

これまでのコマンドは重ねて描くことも可能です。図を出すときにはplt.show()とします。

````python
x = np.linspace(-1, 1 ,10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.quiver(x, y, X, Y, color="blue")
plt.contour(X, Y, Z)

plt.colorbar()
plt.show()
````

## **並べて描く**





### **順番に描く**

まずは、上から順に描いてゆくやり方です。

```python
plt.figure(figsize=(横長, 縦長))

plt.subplot(縦ブロックの数, 横ブロックの数, i)
i番目のプロット処理
```


````python
x = np.linspace(.1, 10)

plt.figure(figsize=(14,4))

plt.subplot(2,3,1)
plt.plot(x, np.sin(x))

plt.subplot(2,3,2)
plt.plot(x, np.cos(x))

plt.subplot(2,3,3)
plt.plot(x, np.tan(x))

plt.subplot(2,3,4)
plt.plot(x, np.exp(x))

plt.subplot(2,3,5)
plt.plot(x, np.log(x))

plt.subplot(2,3,6)
plt.plot(x, np.tanh(x))

````

### **オブジェクト指向的に描く**

```python
fig = plt.figure(figsize=(横長, 縦長))

ax = fig.add_subplot(縦ブロックの数, 横ブロックの数, i)
ax.プロット処理
```

各プロットのブロックを制御するaxisオブジェクト（ax）を作って、そのメソッドでそれぞれのグラフをプロットする方法です。それぞれのaxisを別名で定義しておけば、プロット処理するのに順序は関係ありません。

````python
x = np.linspace(.1, 10)

fig = plt.figure(figsize=(14,4))

ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)
ax5 = fig.add_subplot(2,3,5)
ax6 = fig.add_subplot(2,3,6)

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))
ax4.plot(x, np.exp(x))
ax5.plot(x, np.log(x))
ax6.plot(x, np.tanh(x))
````
